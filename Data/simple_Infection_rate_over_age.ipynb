```python

import numpy as np
from numpy.polynomial import Polynomial

x = np.array([ 8, 24, 35, 45, 55, 65, 75, 85 ])
y = np.array([0.000021, 0.000102, 0.000210, 0.001010,
              0.003950, 0.014300, 0.046100, 0.150000])

log_y = np.log(y)
coeffs = np.polyfit(x, log_y, deg=3)
poly_3 = np.poly1d(coeffs)
def ifr_age(age):
    return np.exp(poly_3(age))
```

```python
import matplotlib.pyplot as plt


x = np.array([ 8, 24, 35, 45, 55, 65, 75, 85 ])
y = np.array([0.000021, 0.000102, 0.000210, 0.001010,
              0.003950, 0.014300, 0.046100, 0.150000])

log_y = np.log(y)

coeffs = np.polyfit(x, log_y, deg=3)
poly_3 = np.poly1d(coeffs)

def ifr_age(age):
    return np.exp(poly_3(age))

x_fit = np.linspace(min(x), max(x), 100)
y_fit = ifr_age(x_fit)

plt.plot(x, y, 'o', label='Data')
plt.plot(x_fit, y_fit, '-', label='Polynomial Fit')
plt.xlabel('Age')
plt.ylabel('IFR')
plt.title('IFR vs. Age with Polynomial Fit')
plt.legend()
plt.show()

print("Polynomial Function (IFR = exp(polynomial)):")
poly_3
```
