```python


# Load full dataset via pagination
def load_full_dataset(dataset_id, page_size=1000):
    """
    Loads the full dataset from the CBS OData API by paging through results.

    Parameters:
      dataset_id (str): The CBS dataset identifier (e.g. "85451NED")
      page_size (int): Number of rows per page.

    Returns:
      DataFrame: All rows concatenated.
    """
    rows = []
    skip = 0
    while True:
        url = f"https://opendata.cbs.nl/ODataApi/odata/{dataset_id}/TypedDataSet?$top={page_size}&$skip={skip}"
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()
        batch = data.get("value", [])
        if not batch:
            break
        rows.extend(batch)
        skip += page_size
        print(f"Loaded {skip} rows for {dataset_id}...")
    return pd.DataFrame(rows)

# Main analysis function
def analyze_cbs_immigration_data():
    print("Loading dataset 85451NED (Immigration flows by nationality)...")
    df_immig_nat = load_full_dataset("85451NED")
    print(f"Dataset 85451NED loaded: {len(df_immig_nat)} rows.\n")

    print("Loading dataset 85371NED (Population born outside the Netherlands)...")
    df_immigrant_pop = load_full_dataset("85371NED")
    print(f"Dataset 85371NED loaded: {len(df_immigrant_pop)} rows.\n")

    print("Loading dataset 85384NED (Population by origin and country of birth)...")
    df_pop_origin = load_full_dataset("85384NED")
    print(f"Dataset 85384NED loaded: {len(df_pop_origin)} rows.\n")

    print("Loading dataset 82025NED (Invited refugees)...")
    df_refugees = load_full_dataset("82025NED")
    print(f"Dataset 82025NED loaded: {len(df_refugees)} rows.\n")

    print("Loading dataset 85568NED (Work accidents by origin and working duration)...")
    df_accidents = load_full_dataset("85568NED")
    print(f"Dataset 85568NED loaded: {len(df_accidents)} rows.\n")

    # Analysis 1: Inhabitant Density (Immigrant vs. Native)
    # From dataset 85371NED, get immigrant population per period.
    immigrant_by_period = (df_immigrant_pop.groupby("Perioden")["BevolkingGeborenBuitenNederland_1"]
                             .sum().reset_index()
                             .rename(columns={"BevolkingGeborenBuitenNederland_1": "ImmigrantPopulation"}))

    # From dataset 85384NED, get total population per period.
    total_by_period = (df_pop_origin.groupby("Perioden")["Bevolking_1"]
                       .sum().reset_index()
                       .rename(columns={"Bevolking_1": "TotalPopulation"}))

    # Merge the two on the period.
    pop_merge = pd.merge(total_by_period, immigrant_by_period, on="Perioden", how="inner")
    pop_merge["NativePopulation"] = pop_merge["TotalPopulation"] - pop_merge["ImmigrantPopulation"]
    pop_merge["ImmigrantProportion"] = pop_merge["ImmigrantPopulation"] / pop_merge["TotalPopulation"]
    pop_merge = pop_merge.sort_values("Perioden")

    # Plot immigrant proportion over time.
    plt.figure(figsize=(10,6))
    plt.plot(pop_merge["Perioden"], pop_merge["ImmigrantProportion"], marker='o', linestyle='-')
    plt.xlabel("Period")
    plt.ylabel("Immigrant Proportion")
    plt.title("Share of Immigrants in the Dutch Population Over Time")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("=== Analysis 1: Inhabitant Density ===")
    print(pop_merge[["Perioden", "NativePopulation", "ImmigrantPopulation", "TotalPopulation", "ImmigrantProportion"]])

    # Analysis 2: Main Flow of Increase in Inhabitants (Migration Flows)
    flow_df = (df_immig_nat.groupby("Perioden")
               .agg({
                   "Immigratie_1": "sum",
                   "EmigratieInclusiefAdministratieveC_2": "sum",
                   "MigratiesaldoInclusiefAdministratie_3": "sum"
               }).reset_index())
    flow_df = flow_df.sort_values("Perioden")

    # Plot net migration over time.
    plt.figure(figsize=(10,6))
    plt.plot(flow_df["Perioden"], flow_df["MigratiesaldoInclusiefAdministratie_3"], marker='o', color='green')
    plt.xlabel("Period")
    plt.ylabel("Net Migration")
    plt.title("Net Migration Flow Over Time")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("=== Analysis 2: Migration Flows ===")
    print(flow_df)

    # Analysis 3: Accidents / Police Encounters (Using Work Accidents as a Proxy, this is of course not right, but it's a good proof of work at least.)
    # Group by period and compute average work accidents.
    accidents_df = (df_accidents.groupby("Perioden")["TotaalWerknemersMetEenOngeval_1"]
                    .mean().reset_index())
    accidents_df = accidents_df.sort_values("Perioden")

    plt.figure(figsize=(10,6))
    plt.plot(accidents_df["Perioden"], accidents_df["TotaalWerknemersMetEenOngeval_1"],
             marker='o', color='red')
    plt.xlabel("Period")
    plt.ylabel("Avg. Work Accidents")
    plt.title("Trend in Work Accidents Over Time")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("=== Analysis 3: Work Accidents Trend ===")
    print(accidents_df)

    # Analysis 4: Additional Conclusions
    # 4a. Cumulative Net Migration
    flow_df["CumulativeNetMigration"] = flow_df["MigratiesaldoInclusiefAdministratie_3"].cumsum()

    plt.figure(figsize=(10,6))
    plt.plot(flow_df["Perioden"], flow_df["CumulativeNetMigration"], marker='o', color='blue')
    plt.xlabel("Period")
    plt.ylabel("Cumulative Net Migration")
    plt.title("Cumulative Net Migration in the Netherlands")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("=== Analysis 4a: Cumulative Net Migration ===")
    print(flow_df[["Perioden", "CumulativeNetMigration"]])

    # Analysis 4: Additional Conclusions (Continued)
    # 4b. Trend in Invited Refugees
    refugees_df = (df_refugees.groupby("Perioden")["UitgenodigdeVluchtelingen_1"]
                   .sum().reset_index())
    refugees_df = refugees_df.sort_values("Perioden")

    plt.figure(figsize=(10,6))
    plt.plot(refugees_df["Perioden"], refugees_df["UitgenodigdeVluchtelingen_1"],
             marker='o', color='purple')
    plt.xlabel("Period")
    plt.ylabel("Number of Invited Refugees")
    plt.title("Trend in Invited Refugees Over Time")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("=== Analysis 4b: Invited Refugees Trend ===")
    print(refugees_df)

    # Final Analytical Conclusions (simple ChatGPT help here)
    print("\nFinal Conclusions:")
    print("1. Inhabitant Density: The graphs above show the share of immigrants relative to the total population over time. "\
          "An increasing immigrant proportion suggests a growing impact on overall population composition.")
    print("2. Migration Flows: The net migration flow, derived from immigration and emigration figures, indicates "\
          "the direct effect of migration on population growth. Positive net migration points to an increasing population due to immigration.")
    print("3. Work Accidents: The trend in work-related accidents (although not broken down by ethnicity in this dataset) "\
          "can be a proxy to further analyze workplace safety and, with additional disaggregation, might shed light on differential risks among groups.")
    print("4. Additional Insights: The cumulative net migration and the trend in invited refugees complement the migration flow data, "\
          "providing further evidence of the immigrant influx and its potential long-term impact on Dutch demographics.")

# Run the analysis
analyze_cbs_immigration_data()
```
