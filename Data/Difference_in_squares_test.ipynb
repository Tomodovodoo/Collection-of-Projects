```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

start = 100
end = 121
steps = 1000

values = np.linspace(start, end, steps)
actual_sqrt = np.sqrt(values)
approx_sqrt = np.sqrt(start) + (values - start) / (np.sqrt(start)*2+1)

differences =  approx_sqrt - actual_sqrt
data = {
    "Value": values,
    "Actual Sqrt": actual_sqrt,
    "Approx Sqrt": approx_sqrt,
    "Difference": differences
}
df = pd.DataFrame(data)

df.to_csv("sqrt_approximation_results.csv", index=False)
print("Results saved to 'sqrt_approximation_results.csv'")

plt.figure(figsize=(10, 6))
plt.plot(values, differences, label="Difference (Actual - Approximation)")
plt.axhline(0, color='red', linestyle='--', linewidth=0.8, label="Zero Difference")
plt.title("Difference Between Actual and Approximate Square Roots")
plt.xlabel("Value")
plt.xticks(ticks=np.linspace(start, end, 5), labels=['0th', '1/4th', '1/2th', '3/4th', '1'])
plt.ylabel("Difference")
plt.legend()
plt.grid()
plt.show()
```

```python
z_values = np.linspace(20, 100, 1000)

def calculate_dmin(z):
    return (z / (2 * z + 1)) + (1 / (8 * z + 4)) - (1 / 2)

dmin_values = calculate_dmin(z_values)
data = {
    "z": z_values,
    "Dmin": dmin_values
}
df = pd.DataFrame(data)

df.to_csv("dmin_results.csv", index=False)
print("Results saved to 'dmin_results.csv'")

plt.figure(figsize=(10, 6))
plt.plot(z_values, dmin_values, label="Dmin", linestyle='-', linewidth=2)
plt.title("Minimum Difference (Dmin) as a Function of z")
plt.xlabel("z")
plt.ylabel("Dmin")
plt.axhline(0, color='red', linestyle='--', linewidth=0.8, label="Zero Line")
plt.legend()
plt.grid()
plt.show()
```
