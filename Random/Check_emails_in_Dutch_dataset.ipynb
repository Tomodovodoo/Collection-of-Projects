```python
from google.colab import files
uploaded = files.upload()

filename = list(uploaded.keys())[0]
with open(filename, 'r') as f:
    lines = f.readlines()

import re
from collections import Counter

email_regex = re.compile(r'[\w\.-]+@[\w\.-]+')

emails = []

for line in lines:
    # Find all email addresses in the line
    matches = email_regex.findall(line)
    emails.extend(matches)

domains = [email.split('@')[1] for email in emails]
domain_counts = Counter(domains)

top_10 = domain_counts.most_common(20)
print("Top 10 most popular domains:")
for domain, count in top_10:
    print(f"{domain}: {count}")

tele2_count = domain_counts.get('tele2.nl', 0)
print(f"\nNumber of @tele2.nl emails: {tele2_count}")
```

```python
top_100 = domain_counts.most_common(100)
print("Top 10 most popular domains:")
for domain, count in top_100:
    print(f"{domain}: {count}")
```
