```python
# Import required packages
import qrcode
from PIL import Image
from IPython.display import display

# Generate QR code
url = "https://www.google.docs.com"
qr = qrcode.make(url)

# Display QR code image
display(qr)

# Optionally save it
qr.save("google_docs_qr.png")
```
