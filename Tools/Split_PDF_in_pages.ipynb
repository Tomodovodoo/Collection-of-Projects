```python

!pip install pymupdf

import fitz  # PyMuPDF
from google.colab import files
import os

def split_pdf(file_path):
    pdf_document = fitz.open(file_path)
    pdf_name = os.path.splitext(os.path.basename(file_path))[0]

    output_dir = pdf_name + '_split'
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for page_num in range(len(pdf_document)):
        pdf_writer = fitz.open()
        pdf_writer.insert_pdf(pdf_document, from_page=page_num, to_page=page_num)

        output_pdf_path = os.path.join(output_dir, f'{pdf_name}_page_{page_num + 1}.pdf')
        pdf_writer.save(output_pdf_path)
        pdf_writer.close()

    return output_dir

uploaded = files.upload()

for file_name in uploaded.keys():
    output_directory = split_pdf(file_name)
    print(f"PDF has been split. The split files are saved in the directory: {output_directory}")
```
