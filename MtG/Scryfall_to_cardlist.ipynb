```python

import requests

# Fetch data from Scryfall API
url = "https://api.scryfall.com/cards/search?q=set:DFT+t:creature"
response = requests.get(url)
data = response.json()

# Extract relevant fields and format them as printed text
output_text = ""
for card in data.get("data", []):
    if "card_faces" in card:  # Handle double-faced cards
        for face in card["card_faces"]:
            card_text = (
                f"{face.get('name', 'Unknown')} {face.get('mana_cost', 'N/A')}\n"
                f"{face.get('oracle_text', 'N/A')}\n"
                f"{face.get('power', 'N/A')}/{face.get('toughness', 'N/A')}\n\n"
            )
            output_text += card_text
    else:
        card_text = (
            f"{card.get('name', 'Unknown')} {card.ge7t('mana_cost', 'N/A')}\n"
            f"{card.get('oracle_text', 'N/A')}\n"
            f"{card.get('power', 'N/A')}/{card.get('toughness', 'N/A')}\n\n"
        )
        output_text += card_text

# Save to a text file
with open("Aetherdrift_Creatures.txt", "w", encoding="utf-8") as file:
    file.write(output_text)

# Print the output
print(output_text)

print("Data saved as 'Aetherdrift_Creatures.txt'")
```
