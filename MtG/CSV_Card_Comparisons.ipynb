```python
from tqdm import tqdm

similarity_scores = {}

# Calculate similarity scores
with tqdm(total=len(owned_cards_list) * len(generated_cards_list)) as pbar:
    for generated in generated_cards_list:
        for owned in owned_cards_list:
            similarity_scores[(generated, owned)] = similar(generated, owned)
            pbar.update(1)
```

```python
extracted_info_list = []

# Iterate over similarity scores and extract information
for (generated_card, owned_card), similarity_ratio in similarity_scores.items():
    if similarity_ratio >= threshold:
        owned_idx = owned_cards_list.index(owned_card)
        generated_idx = generated_cards_list.index(generated_card)

        # Find the Type, Subtype, and Colour information
        extracted_info = cards_list_9_10[cards_list_9_10['name'] == owned_card]
        non_owned_info = generated_cards_df[generated_cards_df['Name'] == generated_card]

        extracted_info_list.append({'Name': owned_card,
                                    'Set': non_owned_info['Set'].values[0],
                                    'Type': extracted_info['Type'].values[0],
                                    'Subtype': extracted_info['Subtype'].values[0],
                                    'Colour': extracted_info['Colour'].values[0],
                                    'List': extracted_info['list'].values[0]})
for (generated_card, owned_card), similarity_ratio in similarity_scores.items():
    if similarity_ratio >= threshold:
        owned_idx = owned_cards_list.index(owned_card)
        generated_idx = generated_cards_list.index(generated_card)

if len(extracted_info_list) == 0:
   extracted_info_list.append({'Name': 'No matches found',
                                            'Set': 'No matches found',
                                            'Type': 'No matches found',
                                            'Subtype': 'No matches found',
                                            'Colour': 'No matches found',
                                            'List': 'No matches found'})

for extracted_info in extracted_info_list:
    with open('owned_cards_extracted_info.csv', 'a', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=['Name', 'Set', 'Type', 'Subtype', 'Colour', 'List'])
        # If the file is empty, write the header
        if file.tell() == 0:
            writer.writeheader()
        writer.writerow(extracted_info)


extracted_info_df = pd.DataFrame(extracted_info_list)
extracted_info_df.to_csv('owned_cards_extracted_info.csv', index=False)

print("Information extracted for owned cards and saved to 'owned_cards_extracted_info.csv'.")
```
